```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test_er.csv', index_col=0)
```

```python
df
plt.figure(figsize=(10, 6))
```

```python
df.loc[100:, '0'].plot()
```

```python
df.loc[500:, '1'].plot()
```

```python
df['returns'] = np.log(np.abs(df['0'])/np.abs(df['0'].shift(1)))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df.loc[100:, 'returns'], label='M_t')
plt.title('Magnetization Over Time')
plt.xlabel('t')
plt.ylabel('Magnetization (M_t)')
plt.legend()
plt.grid(True)
```
